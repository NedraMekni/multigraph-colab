### Let's build adjacency list for nodes in a graph based on information extracted from molecular structures stored in the graph_dict


Data Extraction:
----------------

Retrieves essential information from the graph_dict for the current file. This includes dictionaries for nodes (nodes_p), node entry types (nodes_p_entry), atoms (atom_p), elements (elements_p), and node indices (nodes_i).

NeighborSearch:
---------------

Utilizes the NeighborSearch to conduct neighbor searches on atoms within the molecular structure

Hydrogen Labeling:
-------------------

- Initializes a dictionary (hydrogen_label) to store the count of hydrogen neighbors for each atom.
- Iterates over atoms in the structure (atom_p), determining the number of hydrogen neighbors for each atom.
- Creates a one-hot encoded representation (n_hydro_hot) based on the count of hydrogen neighbors.

Multigraph Generation:
-----------------------

Iterates over nodes in nodes_p:
- For atoms that are not of type 'ATOM', utilizes the NeighborSearch object to find adjacent nodes with different search radii.
- Adds edges to the my_edges list based on the indices of the nodes and their neighbors, contributing to the construction of a multigraph's adjacency list.
